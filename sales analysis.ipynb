```python
# Convert Ship_Date to datetime, handling errors
Sales_Performance['Ship_Date_Format'] = pd.to_datetime(Sales_Performance['Ship_Date'], errors='coerce')

# Check for invalid entries (NaT)
if Sales_Performance['Ship_Date_Format'].isna().any():
    print("Invalid Ship_Date entries found. Dropping them.")
    Sales_Performance = Sales_Performance.dropna(subset=['Ship_Date_Format'])

# Replace original Ship_Date column
Sales_Performance['Ship_Date'] = Sales_Performance['Ship_Date_Format']

# Drop the temporary column (optional)
Sales_Performance.drop(columns=['Ship_Date_Format'], inplace=True)

# Confirm the changes
print(Sales_Performance[['Ship_Date']].head())
```

```python
Regional_Sales= Sales_Performance.groupby(Sales_Performance['Region'])['Sales'].sum().reset_index()
Regional_Sales
```

```python
Yearly_Sales.plot(x='Order_Date',y='Sales', figsize=(8,6), marker='o')
plt.title('Year Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
Regional_Sales.plot(x='Region',y='Sales', kind='bar',figsize=(8,6),width=0.3)
plt.title('Regional Sales Performance')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'Category': ['Furniture', 'Office Supplies', 'Technology'],
    'Sales': [50000, 30000, 20000]
}
Product_Category_Sales = pd.DataFrame(data)

# Plot the pie chart
Product_Category_Sales.plot(
    y='Sales',
    kind='pie',
    figsize=(5, 5),
    autopct='%1.1f%%',
    legend=False,
    labels=Product_Category_Sales['Category']
)

# Add title and display the plot
plt.title('Product Category Sales Distribution')
plt.ylabel('')  # Optional: Remove the default y-axis label
plt.show()
```
